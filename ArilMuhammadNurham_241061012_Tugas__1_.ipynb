```python
# 1. Membuat data JSON manual (format List of Dictionaries)
data_json = """
[
  {"id_produk": 101, "nama_barang": "Laptop Asus", "kategori": "Elektronik", "stok": 15, "harga": 8500000},
  {"id_produk": 102, "nama_barang": "Mouse Wireless", "kategori": "Aksesoris", "stok": 50, "harga": 150000},
  {"id_produk": 103, "nama_barang": "Monitor LED 24", "kategori": "Elektronik", "stok": 8, "harga": 2100000},
  {"id_produk": 104, "nama_barang": "Keyboard RGB", "kategori": "Aksesoris", "stok": 25, "harga": 450000},
  {"id_produk": 105, "nama_barang": "Webcam HD", "kategori": "Elektronik", "stok": 12, "harga": 600000}
]
"""

# Mengonversi string JSON menjadi objek Python
data_csv = json.loads(data_json)
```
